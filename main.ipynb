# Train / Test 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['Daily Time Spent on Site','Age','Area Income', 'Daily Internet Usage','Male']]
y = data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

# 로지스틱 리그레션 모델 만들기

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

# 예측 및 평가

```python
pred = model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

accuracy가 0.9로 굉장히 높은 예측값을 보여줬지만 실전에서 너무 높은 예측값이 나온다면 validation을 여러차례 더 시도 할 꺼 같습니다.

```python
confusion_matrix(y_test, pred)
```

FP보다 FN이 더 높은 오답률을 보여주고 있습니다. 둘 다 낮추는 것도 중요하지만 false유저에게 비용을 더 투자하는 것을 피하는 것보다 positive유저를 한 명이라도 더 정확하게 맞추는 것을 목표로 모델링 해야할 꺼 같습니다.
